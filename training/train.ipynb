```python
# ============================================================
# BrailleBridge — Complete Training Cell
# Paste this entire block. Run once. Takes ~35-40 minutes.
# ============================================================

# STEP 1 — Install libraries
print("Installing libraries...")
import subprocess
subprocess.run(["pip", "install", "roboflow", "ultralytics", "-q"])
print("✅ Libraries installed")

# STEP 2 — Verify GPU
import torch
gpu_available = torch.cuda.is_available()
print(f"GPU available: {gpu_available}")
if not gpu_available:
    print("⛔ STOP — Change runtime to T4 GPU first")
    raise SystemExit("No GPU found")
print(f"GPU: {torch.cuda.get_device_name(0)}")

# STEP 3 — Download dataset
print("\nDownloading dataset...")
from roboflow import Roboflow

# PASTE YOUR REAL API KEY BELOW (from Roboflow Settings)
rf = Roboflow(api_key="uTMToln3SuLjdXwXaurE")
project = rf.workspace("janvis-workspace-rmimv").project("braille-detection-v2-tuhrd")
version = project.version(1)
dataset = version.download("yolov8")

import os
yaml_path = os.path.join(dataset.location, "data.yaml")
print(f"✅ Dataset downloaded at: {dataset.location}")
print(f"✅ data.yaml exists: {os.path.exists(yaml_path)}")

# STEP 4 — Train model
print("\nStarting training — takes ~35 minutes...")
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
results = model.train(
    data=yaml_path,
    epochs=30,
    imgsz=640,
    device=0,
    project="/content/braillebridge",
    name="run1",
    patience=10,
    save=True,
    exist_ok=True
)
print("✅ Training complete!")

# STEP 5 — Verify all files exist
print("\nVerifying output files...")
files_to_check = [
    "/content/braillebridge/run1/weights/best.pt",
    "/content/braillebridge/run1/weights/last.pt",
    "/content/braillebridge/run1/results.png",
    "/content/braillebridge/run1/confusion_matrix.png",
]
for f in files_to_check:
    exists = os.path.exists(f)
    print(f"{'✅' if exists else '❌'} {f}")

print(f"\n✅ data.yaml at: {yaml_path}")
print("\n🎉 ALL DONE — Now download files using the code in the next cell")
```
